```python
def get_idx_from_instagram(index, posts):
    account_id, post_id = posts.loc[index, ['account_id', 'post_id']]
    account_id = str(account_id)
    post_id = str(post_id)
    
    return account_id, post_id
```

```python
for idx in instagram.index.values:
    url, post_type = instagram.loc[idx, ['URL', 'Type']]
    print(url, post_type)
```

```python
def get_images_videos(idx, instagram):
    url, post_type = instagram.loc[idx, ['URL', 'Type']]
    page_link = url
    res = requests.get(page_link, timeout=5)
    html = res.text
    
    soup_image = BeautifulSoup(html, 'html.parser')
    image = soup_image.find("meta", property="og:image")
    print("image: %s " % image["content"])
    image_url = image["content"]

    img_file = "./instagram/instagram_images/" + '{:03d}'.format(idx+1) + ".jpg"
    print(img_file)
    file = open(img_file, 'wb')
    file.write(urllib.request.urlopen(image_url).read())
    
    file.close()
    
    if post_type == 'video':
        soup_video = BeautifulSoup(html, 'html.parser')
        video = soup_video.find("meta", property="og:video")
        #print(video)
        print("video: %s " % video["content"])
        video_url = video["content"]

        vid_file = "./instagram/instagram_videos/" + '{:03d}'.format(idx+1) + ".mp4"
        file = open(vid_file, 'wb')
        file.write(urllib.request.urlopen(video_url).read())
        file.close()
```

```python
get_images_videos(1, instagram)
```

```python
for idx in instagram.index.values:
    get_images_videos(idx, instagram)
```
